# Husky (타겟보드 : CW312) 데이터 쓰기 & 읽기 테스트

```python
MAX_DATA_LEN = 245  # 245이하 크기 설정
random.seed(1)

data_k = bytearray(random.randint(0, 255) for _ in range(MAX_DATA_LEN))
data_p = bytearray(random.randint(0, 255) for _ in range(MAX_DATA_LEN))

print('\n입력 데이터')
print(f'input data_k : {data_k.hex(" ")} ')  
print(f'input data_p : {data_p.hex(" ")} ') 
```

```python
print('\n타겟보드 시리얼버퍼 초기화, 변수들 주입(simpleserial_main/simpleserial-base.c  ->  my_init)') 
print('data_k:') 
target.flush()
target.send_cmd(cmd=0x81, scmd=ord('k'), data=data_k)
ret_k = target.read_cmd(timeout=2000)  
print(f'read data_k chr  : {chr(ret_k[1])} ')
print(f'read data_k len  : {ret_k[2]} ')
print(f'read data_k data : {ret_k[3:3+ret_k[2]].hex(" ")} ')
print(f'read data_k crc  : {(hex(ret_k[-2]))} ')
```

```python
print('data_p:') 
target.flush()
target.send_cmd(cmd=0x81, scmd=ord('p'), data=data_p)
ret_p = target.read_cmd(timeout=2000)         
print(f'read data_p chr  : {chr(ret_p[1])} ')
print(f'read data_p len  : {ret_p[2]} ')
print(f'read data_p data : {ret_p[3:3+ret_p[2]].hex(" ")} ')
print(f'read data_p crc  : {hex(ret_p[-2])} ')
```

```python
print('length of return data:') 
target.flush()
target.send_cmd(cmd=0x81, scmd=ord('l'), data=bytearray([MAX_DATA_LEN]))
ret_len = target.read_cmd(timeout=2000)         
print(f'read data_p chr  : {chr(ret_len[1])} ')
print(f'read data_p len  : {ret_len[2]} ')
print(f'read data_p data : {ret_len[3:3+ret_len[2]][0]} ')
print(f'read data_p crc  : {hex(ret_len[-2])} ')
```

```python
print('\n예제 연산(simpleserial_main/simpleserial-base.c  ->  my_update)') 
target.flush()
target.send_cmd(cmd=0x82, scmd=ord('c'), data=[])
```

```python
print('\n결과 출력(simpleserial_main/simpleserial-base.c  ->  my_final)') 
print('ret_k_XOR_p:') 
target.flush()
target.send_cmd(cmd=0x83, scmd=ord('r'), data=[])
ret_k_XOR_p = target.read_cmd(timeout=2000)         
print(f'read ret_k_XOR_p chr  : {chr(ret_k_XOR_p[1])} ')
print(f'read ret_k_XOR_p len  : {ret_k_XOR_p[2]} ')
print(f'read ret_k_XOR_p data : {ret_k_XOR_p[3:3+ret_k_XOR_p[2]].hex(" ")} ')
print(f'read ret_k_XOR_p crc  : {hex(ret_k_XOR_p[-2])} ')
```

```python
print('\n결과 비교') 
Return_k_XOR_p = ret_k_XOR_p[3:3+ret_k_XOR_p[2]]
Golden_k_XOR_p = bytes(x ^ y for x, y in zip(data_k, data_p))

print(f'Return_k_XOR_p : {Return_k_XOR_p.hex(" ")}')
print(f'Golden_k_XOR_p : {Golden_k_XOR_p.hex(" ")}')


print(Golden_k_XOR_p == Return_k_XOR_p)
```
